```python
# Importar la biblioteca Pandas para la manipulación de datos
import pandas as pd

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv("/content/country_vaccinations.csv")

# Mostrar las primeras filas del DataFrame para visualizar la estructura de los datos
df.head()
```

```python
# Mostrar información del DataFrame: número de columnas, tipos de datos y valores nulos
df.info()

# Convertir la columna 'date' a formato datetime64 para facilitar operaciones con fechas
df['date'] = pd.to_datetime(df['date'])
```

```python
# Agrupar los datos por tipo de vacuna y sumar las vacunas aplicadas en cada caso
vacunas_por_compania = df.groupby('vaccines')['daily_vaccinations'].sum()

# Mostrar el resultado
print(vacunas_por_compania)
```

```python
# Calcular el total de vacunas aplicadas en el mundo sumando la columna 'daily_vaccinations'
total_vacunas = df['daily_vaccinations'].sum()

# Mostrar el resultado
print("Total vacunas aplicadas en el mundo:", total_vacunas)
```

```python
# Agrupar por país y calcular el promedio de vacunas aplicadas diariamente por pais
promedio_por_pais = df.groupby("country")["daily_vaccinations"].mean()

# Mostrar el resultado
print(promedio_por_pais)
```

```python
# Filtrar el DataFrame para obtener solo los registros del 29 de enero de 2021
vacunas_29ene = df[df['date'] == '2021-01-29']['daily_vaccinations'].sum()

# Mostrar el total de vacunas aplicadas en esa fecha
print("Vacunas aplicadas el 29/01/21:", vacunas_29ene)
```

```python
# Crear una nueva columna 'diferencias' con la diferencia entre vacunaciones diarias reportadas y sin procesar
df['diferencias'] = df['daily_vaccinations'] - df['daily_vaccinations_raw']

# Crear un nuevo DataFrame con la columna derivada
conDiferencias = df.copy()

# Mostrar las primeras filas del nuevo DataFrame
conDiferencias.head()
```

```python
# Obtener la fecha más reciente en el conjunto de datos
fecha_max = df['date'].max()

# Obtener la fecha más antigua en el conjunto de datos
fecha_min = df['date'].min()

# Calcular la diferencia de tiempo entre ambas fechas
periodo_tiempo = fecha_max - fecha_min

# Mostrar el resultado
print("Periodo de tiempo total:", periodo_tiempo)
```

```python
# Crear un nuevo DataFrame con la cantidad de vacunas aplicadas en cada registro
df['canVac'] = df['daily_vaccinations']

# Guardar la versión con la nueva columna
conCantidad = df.copy()

# Mostrar las primeras filas del nuevo DataFrame
conCantidad.head()
```

```python
# Filtrar registros con fecha antes del 20 de diciembre de 2020
antes20 = df[df['date'] < '2020-12-20']

# Mostrar las primeras filas del nuevo DataFrame
antes20.head()
```

```python
# Filtrar registros de vacunación que mencionen Pfizer en la columna 'vaccines'
pfizer = df[df['vaccines'].str.contains("Pfizer", na=False)]

# Mostrar las primeras filas del nuevo DataFrame
pfizer.head()
```

```python
# Guardar cada DataFrame en archivos CSV para su análisis posterior
conDiferencias.to_csv("conDiferencias.csv", index=False)
conCantidad.to_csv("conCantidad.csv", index=False)
antes20.to_csv("antes20.csv", index=False)
pfizer.to_csv("pfizer.csv", index=False)
```

```python
# Extraer la cantidad de vacunas utilizadas cada día
# Separamos los nombres de vacunas en la columna 'vaccines' por el carácter ","
df['canVac'] = df['vaccines'].str.split(',').apply(len) * df['daily_vaccinations']

# Crear un nuevo DataFrame con la columna derivada 'canVac'
conCantidad = df.copy()

# Mostrar las primeras filas del DataFrame generado
conCantidad.head()
```

```python
# Filtrar los registros con fecha anterior al 20 de diciembre de 2020
antes20 = df[df['date'] < '2020-12-20']

# Mostrar las primeras filas del DataFrame generado
antes20.head()
```

```python
# Filtrar registros donde la columna 'vaccines' contenga la palabra 'Pfizer'
pfizer = df[df['vaccines'].str.contains("Pfizer", na=False)]

# Mostrar las primeras filas del DataFrame generado
pfizer.head()
```

```python
# Exportar los DataFrames generados a un archivo Excel con múltiples hojas
with pd.ExcelWriter("resultadosReto.xlsx") as writer:
    conDiferencias.to_excel(writer, sheet_name="conDiferencias", index=False)
    conCantidad.to_excel(writer, sheet_name="conCantidad", index=False)
    antes20.to_excel(writer, sheet_name="antes20", index=False)
    pfizer.to_excel(writer, sheet_name="pfizer", index=False)
```
